```python
#Renaming the features 
df = df.rename({'R&D Spend':'RDS','Administration':'Admin','Marketing Spend':'MS','Profit':'PFT'},axis = 1)
df.head()
```

## Data Transformation

```python
# state is categorical data so we convert this to numerical data.
# We use here on hot encoding.
df2=pd.get_dummies(df,columns=['State'],drop_first =True)
df2.head()
```

```python
#correlation
df2.corr()
```

## Preparing a Model

```python
# Building our first model
model_1 = smf.ols("PFT~RDS+Admin+MS",data = df2).fit()
```

```python
# Coefficients of the variables
model_1.params
```

```python
#cumulating our summary on model_1
model_1.summary()
```

## Calculating VIF

```python
#calculating ViF values for independent variables
#collinearity check

rsq_RDS = smf.ols("RDS~Admin+MS",data = df2).fit().rsquared
vif_RDS = 1/(1-rsq_RDS)

rsq_Admin = smf.ols("Admin~RDS+MS",data = df2).fit().rsquared
vif_Admin = 1/(1-rsq_Admin)

rsq_MS = smf.ols("MS~Admin+RDS",data = df2).fit().rsquared
vif_MS = 1/(1-rsq_MS)
```

```python
#storing VIF values in a Dataframe 
d1 = {'Variables':['RDS','Admin','MS'],'VIF':[vif_RDS,vif_Admin,vif_MS]}
vif_Frame = pd.DataFrame(d1)
vif_Frame
```

```python
# Here we consider all the variables because vif > 5.
```

### Plotting the Q-Q Plot

```python
#checking for normality of  residuals by qq plot 
st.qqplot (model_1.resid,line = 'q')
plt.title("Normal Q-Q plot of residuals")
py.show()
```

```python
# Checking homoscedacity ot heteroscedacity 
def standard_values(vals):
    return (vals-vals.mean())/vals.std()
```

```python
#residual scatter plot 
plt.scatter(standard_values(model_1.fittedvalues),standard_values(model_1.resid))
plt.title('Residual plot')
plt.xlabel('standardized fitted values')
plt.ylabel('standardized residual values')
plt.show()
```

## Model Deletion Diagnostics

```python
# Checking for outliers or influencers via using Cook's Distance
(d,_)= model_1.get_influence().cooks_distance
d
```

```python
#plot the influencers using the stem plot 
fig = plt.figure(figsize = (20,7))
plt.stem(np.arange(len(df2)),np.round(d,3))
plt.xlabel('row index')
plt.ylabel('Cooks Distance')
plt.show()
```

```python
#locating the value that influencing our plot 
np.argmax(d),np.max(d)
```

```python
#computing out leverage value using high influence plot: points beyond our leverge_cutoff value are outliers
influence_plot(model_1)
plt.show()
```

```python
# Leverage cutoff value is 3(k+1)/n
k = df2.shape[1]
n = df2.shape[0]
leverage_cutoff = 3*((k+1)/n)
leverage_cutoff
```

```python
#locating the influencer in our dataset
df2[df2.index.isin([48,49])]
```

```python
# Influencer dropping the influencer and reset the dataset
df3 = df2.drop(df.index[[48,49]],axis = 0).reset_index(drop = True)
df3
```
