```python
import pandas as pd
import numpy as np
import os
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.formula.api as sm
from statsmodels.graphics.regressionplots import influence_plot
import statsmodels.api as sma
```

## Exploratory Data Analysis

```python
# since doors and gears have same values we drop it 
df3 =  df2.drop(['Doors','Gears'],axis = 1)
df3.head()
```

### Correlation

```python
df3.corr()
```

## Visualization

```python
# here we scatter the plot for all indepedent and dependent variables.
sns.set_style(style = 'darkgrid')
sns.pairplot(df3)
```

```python
# Renaming the columns
df4=df3.rename({'Quarterly_Tax':'QT','Age_08_04':'Age'},axis=1)
```

```python
df4.head()
```

## Building our model

```python
# building our first model without transformaion.
model_1 = sm.ols("Price~Age+Weight+HP+cc+QT",data = df4).fit()
```

```python
# Here we check the parameters of the model.
model_1.params
```

```python
# Here we check the accuracy of our model with summary.
model_1.summary()
```

## Computing VIF

```python
#computing vif values for all are independent variables 
rsq_wt = sm.ols("Weight~QT+HP+Age+cc",data = df4).fit().rsquared
vif_wt = 1/(1-rsq_wt)

rsq_qt = sm.ols("QT~Weight+HP+Age+cc",data = df4).fit().rsquared
vif_qt = 1/(1-rsq_qt)

rsq_hp = sm.ols("HP~Weight+QT+Age+cc",data = df4).fit().rsquared
vif_hp = 1/(1-rsq_hp)

rsq_age = sm.ols("Age~Weight+HP+QT+cc",data = df4).fit().rsquared
vif_age = 1/(1-rsq_age)

rsq_cc = sm.ols("cc~Age+HP+QT+Weight",data = df4).fit().rsquared
vif_cc = 1/(1-rsq_cc)
```

```python
# Here we use Data Frame
t1 = {'Variables':['Weight','QT','HP','cc','Age'],'VIF':[vif_wt,vif_qt,vif_hp,vif_age,vif_cc]}
vif_frame = pd.DataFrame(t1)
vif_frame
```

```python
# Checking for normality of residuals 
sma.qqplot(model_1.resid,line = 'q')
plt.title("Normal Q-Q plot of residuals")
plt.show()
```

```python
# We are locating our residual errors
list(np.where(model_1.resid >6000))
```

```python
# Locating our residual errors
list(np.where(model_1.resid < -6000))
```

```python
# Checking for homoscedacity or heteroscedacity of graph
def standard_values(vals): return(vals-vals.mean())/vals.std()
```

```python
# Residual Scatter Plot
plt.scatter(standard_values(model_1.fittedvalues),standard_values(model_1.resid))
plt.title('residual plot')
plt.xlabel('standardized fitted values')
plt.ylabel('standardized residual values')
plt.show()
```

## Model Deletion Diagnostics

```python
# Model deletion technique by influence plot and leverage values 
# Checking for outliers and influencers via Cook's Distance.
(t,_)= model_1.get_influence().cooks_distance
t
```

```python
# We plot here the stem plot for influencers.
plt.figure(figsize = (20,7))
plt.stem(np.arange(len(df4)),np.round(t,5))
plt.xlabel('row index')
plt.ylabel('Cooks Distance')
plt.show()
```

```python
 np.argmax(t),np.max(t)
```

```python
# Calculating the Leverage value using high influence plot: points beyond our leverge_cutoff value are outliers
influence_plot(model_1)
plt.show()
```
